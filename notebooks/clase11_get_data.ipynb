```python
import re
import json
import random
import pandas as pd

import tweepy
from tweepy import Stream
from tweepy.streaming import StreamListener
```

```python
# Load Keys
keys = pd.read_csv("../archivos/no/keys.csv",header=None)
keys = dict(zip(keys[0],keys[1]))

# Authenticate through the API
auth = tweepy.OAuthHandler(keys['api_key'], keys['api_secret_key'])
auth.set_access_token(keys['access_token'], keys['access_token_secret'])
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
ciudades = pd.read_csv('archivos/co_small.csv')
ciudades
```

# Per city

```python
all_tweets = []

for ciudad in ciudades.iterrows():
    c = ciudad[1]
    gc = f"{c.lat},{c.lng},45mi"
    
    try:

        tweets = tweepy.Cursor(api.search, 
                           q=['día de la madre -RT'], 
                           lang="es",
                           since='2021-03-28', 
                           geocode=gc,
                           tweet_mode='extended',
                           include_rts=False).items(100)


        t = []
        i = 0
        for tweet in tweets:
            i = i+1
            print(i)
            t.append([
                tweet.full_text,
                tweet.created_at,
                c.city,
                c.lat,
                c.lng,
                tweet.user.location,
                tweet.place,
                tweet.coordinates,
                f"https://twitter.com/i/web/status/{tweet.id}"])

        all_tweets.extend(t)
        print(c.city)
        
    except tweepy.TweepError as e:
        print(c.city, e)
```

```python
data = pd.DataFrame(all_tweets)
data.columns = ['tweet', 'date', 'city', 'lat', 'lon', 'user_location', 'place', 'coordinates', 'link']
data
```

```python
tweets = tweepy.Cursor(api.search,
                       q=['machine learning -RT'],
                       lang='en',
                       tweet_mode='extended',
                       include_rts=False).items(100)
```

```python
t = [tweet for tweet in tweets]
```

```python
t_processed = []

for tweet in t:
    temp = []
    temp.append(tweet.full_text)
    temp.append(tweet.created_at)
    temp.append(f"https://twitter.com/i/web/status/{tweet.id}")
    
    t_processed.append(temp)
```

```python
data = pd.DataFrame(t_processed)
data.columns = ['tweet', 'date', 'link']
data.head()
```

```python
data.to_csv("machine_learning.csv", index=False)
```
