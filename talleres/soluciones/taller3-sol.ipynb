# Punto 1:

- `[18 pts]` Hacer Web Scraping de 10 animales en Wikipedia (en búcle)
- `[12 pts]` Obtener el **encabezado** de cada animal
- `[20 pts]` Obtener todos los **textos** que están en las etiquetas de negrilla del primer párrafo.

```python
import bs4 as bs 
import urllib.request
```

```python
mi_lista = ["https://es.wikipedia.org/wiki/Inia_geoffrensis",
            "https://es.wikipedia.org/wiki/Delphinidae",
            "https://es.wikipedia.org/wiki/Felis_silvestris_catus",
            "https://es.wikipedia.org/wiki/Giraffa_camelopardalis",
            "https://es.wikipedia.org/wiki/Struthio_camelus",
            "https://es.wikipedia.org/wiki/Spheniscidae",
            "https://es.wikipedia.org/wiki/Oryctolagus_cuniculus",
            "https://es.wikipedia.org/wiki/Folivora",
            "https://es.wikipedia.org/wiki/Ailuropoda_melanoleuca",
            "https://es.wikipedia.org/wiki/Ornithorhynchus_anatinus",
            ]
```

```python
todas_negrillas = []

for i, enlace in enumerate(mi_lista):
    print(f"Enlace #{i+1:<2}: {enlace}")
    
    # Obtener página web en Python
    request = urllib.request.urlopen(enlace)
    source = request.read() 
    request.close()     
    soup = bs.BeautifulSoup(source,'html.parser') 
    
    # Encabezado
    titulo = soup.find('h1').text
    print(f"Título: {titulo}")
    
    # Primer párrafo
    primer_parrafo = soup.find("p")
    # print(f"Primer párrafo: {primer_parrafo}")
    
    # Iterar sobre los textos en negrilla
    print("\n>>>> NEGRILLA")
    for negrilla in primer_parrafo.find_all('b'):
        print(f"{negrilla.text}")
        todas_negrillas.append(negrilla.text)
    
    print("*"*100)
    print()
```

```python
todas_negrillas = []

for i, enlace in enumerate(mi_lista):
    print(f"Enlace #{i+1:<2}: {enlace}")
    
    # Obtener página web en Python
    request = urllib.request.urlopen(enlace)
    source = request.read() 
    request.close()     
    soup = bs.BeautifulSoup(source,'html.parser') 
    
    # Encabezado
    titulo = soup.find('h1').text
    print(f"Título: {titulo}")
    
    # Para encontrar el primer párrafo
    for c in soup.find(class_="mw-parser-output").children:
        if c.name == "p":
            primer_parrafo = c
            break
    # print(f"Primer párrafo: {primer_parrafo}")
    
    # Iterar sobre los textos en negrilla
    print("\n>>>> NEGRILLA")
    for negrilla in primer_parrafo.find_all('b'):
        print(f"{negrilla.text}")
        todas_negrillas.append(negrilla.text)
    
    print("*"*100)
    print()
```

# Punto 2:
- `[BONUS 10 pts]` Usando regex, reemplazar todas las caracteres especiales del punto anterior (palabras en negrilla) por un asterisco (¡Ojo, los espacios se quedan!)


```python
import re
```

```python
for palabra in todas_negrillas:
    palabra_mod = re.sub("[^a-zA-Z\s]","*", palabra)
    if palabra != palabra_mod:
        print(f'{palabra:>35} ----- {palabra_mod}')
```

```python
cedulas = ["1.128.248.797", "1128248797", "64.897.456", "64897456", "1.1.11.111"]
```

```python
for cedula in cedulas:
    if re.match(r"\d{0,1}.?\d{2,3}.?\d{3}.?\d{3}", cedula): # Hay que ver otros casos
        print(f"{cedula} es una cédula")
```
